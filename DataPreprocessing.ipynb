# Part 1: Combine

```python
import glob
import csv
import os.path
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
with open('output.csv', 'w', newline='') as f_output:
    csv_output = csv.writer(f_output)

    for fname in glob.glob(r'./Companies/*.csv'):
        with open(fname, newline='') as f_input:
            csv_input = csv.reader(f_input)
            fname = os.path.split(fname)[1].split('.')[0]
            
            for i, row in enumerate(csv_input):
                if i !=0:  
                    row.insert(1, fname)
                    csv_output.writerow(row)
```

```python
df = pd.read_csv(r'./output.csv', header=None)
df.columns = ['Date','Company','Employee_title','location','employee_status','review_title','years_at_company','helpful',
              'pros', 'cons', 'advice_to_mgmt', 'rating_overall','rating_balance','rating_culture','rating_career','rating_comp','rating_mgmt']

df.to_csv('all_company.csv')
```
